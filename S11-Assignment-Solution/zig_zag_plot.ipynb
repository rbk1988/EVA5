```
import math
import matplotlib.pyplot as plt

def plot_lr_zigzag(lr_min, lr_max, step_size, num_iterations):
  """."""
  lr = []
  for iteration in range(0,num_iterations+1):
    cycle = math.floor(1 + iteration/(2*step_size))
    x = iteration/(step_size)-2*cycle + 1
    x = x if x>=0  else -1*x 
    lr.append(lr_min + (lr_max - lr_min)*(1 - x))

  plt.plot([x for x in range(0,num_iterations+1)], lr)
  plt.xlabel("x")
  plt.ylabel("y")
```

```
lr_min = 1E-3
lr_max = 1.0
step_size = 80
num_iterations = 1000
plot_lr_zigzag(lr_min, lr_max, step_size, num_iterations)
```
